# Series vs List

```python
           Feature             Python List                     pandas Series
0            Index  Only integer positions             Custom index (labels)
1            Speed   Slower for large data    Faster (uses NumPy internally)
2  Math operations  Not supported directly      Fully vectorized (fast math)
3   Missing values              No support             NaN handling built-in
4          Methods                Very few  Many (mean, sum, describe, etc.)
5      Integration      General Python use            Best for data analysis
```

```python
lst = [10, 20, 30]
s = pd.Series(lst)

print(s * 2)
# Series supports math → list does not.
```

# boolean indexing

```python
#used in filtering the data 
s = pd.Series([10, 20, 30, 40, 50])
s[s > 25]
```

```python
# Filter values greater than 25
filtered = s[s > 25]
print(filtered)
```

```python
s = pd.Series([10, 20, 30, 40, 50])

# Filter values greater than 25
s[s > 25]
```

```python
# Values > 20 AND < 50
filtered = s[(s > 20) & (s < 50)]
print(filtered)
```
