# Using Nested List

```python
import pandas as pd

data = [
    [1, "Renu", 23, "Meerut"],
    [2, "Amit", 25, "Delhi"],
    [3, "Neha", 22, "Noida"]
]
a=["ID", "Name", "Age", "City"]
df = pd.DataFrame(data, columns=a)
print(df)
```

# using list of dict

```python
data = [
    {"ID": 1, "Name": "Renu", "Age": 23, "City": "Meerut"},
    {"ID": 2, "Name": "Amit", "Age": 25, "City": "Delhi"},
    {"ID": 3, "Name": "Neha", "Age": 22, "City": "Noida"}
]

df = pd.DataFrame(data)
print(df)
```

# using dict of list

```python
import pandas as pd    # columnar format

data = {
    "ID": [1, 2, 3],
    "Name": ["Renu", "Amit", "Neha"],
    "Age": [23, 25, 22],
    "City": ["Meerut", "Delhi", "Noida"]
}

df = pd.DataFrame(data)
print(df)
```

```python
# DataFrame from Nested Dictionary (dict inside dict)
```

```python
data = {
    0: {"ID": 1, "Name": "Renu", "Age": 23, "City": "Meerut"},
    1: {"ID": 2, "Name": "Amit", "Age": 25, "City": "Delhi"},
    2: {"ID": 3, "Name": "Neha", "Age": 22, "City": "Noida"}
}

df = pd.DataFrame.from_dict(data, orient='index')
print(df)
```
