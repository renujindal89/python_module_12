```python
#with broadcasting
#NumPy expands b automatically to match A.
A = np.array([[1, 2, 3],
              [4, 5, 6]])

b = np.array([10, 20, 30])

A + b
```

```python
# in python 
result = []
for row in A:
    new_row = []
    for i in range(len(b)):
        new_row.append(row[i] + b[i])
    result.append(new_row)

print(result)
```

# boolean indexing

```python
#NumPy expands b automatically to match A.
A = np.array([[1, 2, 3],
              [4, 5, 6]])

A[A > 3]
# element-wise comparison
```

```python
# in python 
A = [[1, 2, 3],
     [4, 5, 6]]

result = []
for row in A:
    new_row = []
    for x in row:
        new_row.append(x > 3)
    result.append(new_row)

print(result)
```
